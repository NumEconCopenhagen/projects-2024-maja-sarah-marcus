# The relationship between GDP and unemployment

To better show the relationship between GDP and unemployment, we plot them in the same figure. By doing this, we can see that the relationship is seemingly negative.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read GDP and unemployment data from Excel files
GDP_data = pd.read_excel('GDP.xlsx')
Unemployment_data = pd.read_excel('Unemployment.xlsx')

# Set up the figure
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot unemployment development
ax1.plot(Unemployment_data['Year'], Unemployment_data['Unemployed'], marker='s', linestyle='-', color='red', label='Unemployment')
ax1.set_xlabel('Year')
ax1.set_ylabel('Unemployed people', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax1.grid(True)

# Create a twin Axes sharing the xaxis
ax2 = ax1.twinx()

# Plot GDP development on the second y-axis
ax2.plot(GDP_data['Year'], GDP_data['GDP'], marker='o', linestyle='-', color='blue', label='GDP')
ax2.set_ylabel('GDP pr. person (1000 DKK)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Add legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Title
plt.title('Development of GDP and unemployment over time')

plt.tight_layout()
plt.show()
```

The figure titled "Development of GDP and unemployment over time" illustrates the changes in GDP per person (in 1000 DKK) and the number of unemployed people from 2007 to 2023. The horizontal axis represents the years, the left vertical axis (red) shows the number of unemployed people, and the right vertical axis (blue) shows the GDP per person.

The red line with square markers represents the unemployment data, while the blue line with circular markers represents the GDP data.

**Comparative Analysis**\
From 2008 to 2010, there is a sharp increase in unemployment, reaching its peak around 2010. During the same period, GDP shows a decline, hitting its lowest point around 2009. This inverse relationship indicates that during economic downturns, GDP decreases while unemployment increases.

Post-2010, unemployment begins to decline gradually, while GDP starts to recover and shows a steady growth trend, particularly from 2012 onwards. This suggests an economic recovery where increased GDP correlates with reduced unemployment.

Around 2020, there is a notable fluctuation in both curves. Unemployment spikes, while GDP also shows a dip, likely reflecting economic disruptions (such as the COVID-19 pandemic). Following this period, from 2021 onwards, unemployment drops significantly, and GDP increases sharply, indicating a robust economic recovery.

Overall, the figure demonstrates a clear inverse relationship between GDP and unemployment: higher GDP generally correlates with lower unemployment, and vice versa. This visualization effectively highlights the dynamic interplay between economic growth and labor market conditions over the years.

# Finding correlation coefficient between GDP and unemployment

To further specify the relationsship between GDP and unemployment, we find the correlation coefficient between GDP and unemployment.

```python
GDP_values = GDP_data['GDP']
Unemployment_values = Unemployment_data['Unemployed']
correlation = GDP_values.corr(Unemployment_values)
print("correlation coefficient:" , correlation)
```

The displayed code calculates the correlation coefficient between GDP and unemployment. The correlation coefficient is a statistical measure that indicates the strength and direction of the relationship between two variables. In this context, the correlation coefficient is calculated to understand how changes in GDP are associated with changes in unemployment.

The result shows a correlation coefficient of approximately -0.693. This negative value indicates an inverse relationship between GDP and unemployment. In simpler terms, as GDP increases, unemployment tends to decrease, and vice versa. The magnitude of the coefficient (close to -0.7) suggests a moderately strong inverse relationship.

This finding aligns with the economic theory that higher GDP (indicating a stronger economy) typically leads to lower unemployment rates, as more economic activity usually translates to more jobs being available. Conversely, when GDP declines, unemployment tends to rise due to reduced economic activity and job availability. This correlation coefficient quantitatively supports the observed trends in the previous figures, reinforcing the inverse relationship between GDP and unemployment over time.
