# Getting data from Statistics Denmark

Import data for GDP

```python
data = pd.read_excel('GDP.xlsx')
print(data)
```

Descriptive statistics for GDP

```python
plt.figure(figsize=(10, 6))  # Set the size of the figure
plt.plot(data['Year'], data['GDP'], marker='o', linestyle='-')
plt.title('Development of GDP over time')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.grid(True)  # Add gridlines
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to avoid cutting off axis labels
plt.show()
```

Import data for unemployment

```python
data = pd.read_excel('unemployment.xlsx')
print(data)
```

Descriptive statistics for unemployment

```python
plt.figure(figsize=(10, 6))  # Set the size of the figure
plt.plot(data['Year'], data['Unemployed'], marker='o', linestyle='-')
plt.title('Development of unemployed people over time')
plt.xlabel('Year')
plt.ylabel('Unemployed')
plt.grid(True)  # Add gridlines
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to avoid cutting off axis labels
plt.show()
```
